```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
atheletic_sales_2020 = pd.read_csv('../M5_Starter_Code/Resources/athletic_sales_2020.csv')
atheletic_sales_2021 = pd.read_csv('../M5_Starter_Code/Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
atheletic_sales_2020.head()
```

```python
# Display the 2021 sales DataFrame
atheletic_sales_2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
atheletic_sales_2020.dtypes
```

```python
# Check the 2021 sales data types.
atheletic_sales_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_atheletic_sales = pd.concat([atheletic_sales_2020, atheletic_sales_2021])
combined_atheletic_sales = combined_atheletic_sales.reset_index(drop=True)
combined_atheletic_sales.head()
```

```python
# Check if any values are null.
combined_atheletic_sales.isnull()
```

```python
# Check the data type of each column
combined_atheletic_sales.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_atheletic_sales['invoice_date'] = pd.to_datetime(combined_atheletic_sales['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
print(combined_atheletic_sales['invoice_date'].dtype)
```

```python
combined_atheletic_sales.head()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
grouped_df = combined_atheletic_sales.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()
# Rename the sum to "Total_Products_Sold".
grouped_df = grouped_df.rename(columns={'units_sold': 'Total_Products_Sold'})
# Show the top 5 results.
grouped_df.head(5)                    
```
